```python
dow_dict = {
    0:2, # monday
    1:3,
    2:4,
    3:5,
    4:6,
    5:7,
    6:1 # sunday
}

def format_for_model(df):

    out = []

    for _, row in df.iterrows():

        

        ts = row.name

        dow = day_of_week(ts)

        entry = [row["irradiance"], row["temperature"], row["true_power"], ts.hour, dow, ts.month]

        out.append(entry)

    return out

def day_of_week(ts):

    dow_datetime = ts.weekday()

    return dow_dict[dow_datetime]
```
