```python
import numpy as np
import scipy.io as sio
```

```python
path_in="../data/test/input.mat"
path_out="../data/test/output_denorm.mat"
path_pred="../data/test/pred_denorm.mat"
```

```python
def load_first_array(path):
    d = sio.loadmat(path)
    # find first ndarray in the dict (skip meta keys)
    for k, v in d.items():
        if k.startswith("__"):
            continue
        if isinstance(v, np.ndarray):
            return v
    raise ValueError(f"No array found in {path}")
```

```python
in_arr = load_first_array(path_in) 
```

```python
ou_arr = load_first_array(path_out) 
```

```python
p_arr = load_first_array(path_pred) 
```

```python
import pandas as pd
```

```python

def to_pd(arr):
    flattened_data = []
    for sub_array in arr:
        flattened_data.append(sub_array[0])
    
    flattened_data = np.vstack(flattened_data)
    
    return pd.DataFrame(flattened_data, columns=['t-2', 't-1'])
```

```python
in_df = to_pd(in_arr)
in_df
```

```python
ou_df = pd.DataFrame(ou_arr, columns=['real_pv'])
```
